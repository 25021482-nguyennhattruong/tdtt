#### Bài 1
#### Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. 
#### def Max_Of_Two( a, b) -> int:

```python
def max_of_two(a, b):
    if a > b:
        return a
    elif a < b:
        return b
    else:
        return "Hai số bằng nhau"
    
try:
    a, b = map(int, input("Nhập 2 số nguyên (mỗi số cách nhau 1 dấu cách): ").split())
    print(f"Số lớn hơn trong hai số: {max_of_two(a, b)}")
except ValueError:
    print("Vui lòng chỉ nhập số nguyên")
```

#### Bài 3
#### Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def so_nguyen_to(n):
    if n <= 1:
        return f"{n} không phải số nguyên tố"
    else:
        for i in range(2, int(n ** 0.5) + 1):
            if n % i == 0:
                return f"{n} không phải số nguyên tố"
        return f"{n} là số nguyên tố"
    
a = int(input("Nhập số nguyên dương: "))
print(so_nguyen_to(a))
```

#### Bài 4
#### Viết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def so_hoan_hao(n):
    tong = 0
    for i in range(1, n // 2 + 1):
        if n % i == 0:
            tong += i
    return tong == n

a = int(input("Nhập số nguyên dương: "))
if so_hoan_hao(a):
    print(f"{a} là số hoàn hảo")
else:
    print(f"{a} không phải số hoàn hảo")
```

#### Bài 7
#### Viết hàm giả lập một máy tính bỏ túi. Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử \( operat \)
#### Biết các toán tử đơn giản thực hiện phép toán lần lượt là:
#### Phép trừ \( - \)
#### Phép cộng \( + \)
#### Phép chia \( / \)
#### Phép nhân \( * \)
#### Kết quả làm tròn đến chữ số thập phân thứ hai.
#### Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)).
#### Đầu ra: Kết quả phép toán.

```python
def may_tinh(a, b, op):
    ket_qua = 0
    if op == '+':
        ket_qua = a + b
    elif op == '-':
        ket_qua = a - b
    elif op == '*':
        ket_qua = a * b
    elif op == '/':
        if b != 0:
            ket_qua = a / b
        else:
            return "Không thể chia cho 0"
    else:
        return "Toán tử không hợp lệ"
    return round(ket_qua, 2)

a, b = map(float, input("Nhập 2 số thực (mỗi số cách nhau 1 dấu cách): ").split())
op = input("Nhập toán tử: ")
print(f"Kết quả của phép tính {a} {op} {b}: {may_tinh(a, b, op)}")
```

#### Bài 8
#### Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.
#### Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này.

```python
def hamming_distance_fast(x, y):
    return (x ^ y).bit_count()

x, y = map(int, input("Nhập 2 số nguyên dương (mỗi số cách nhau 1 dấu cách): ").split())
print(hamming_distance_fast(x, y))
```

#### Bài 10
#### Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành 'z', 'b' thành 'b' và 'c' thành 'x'.
#### Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai từ đầu vào có cùng độ dài. 


```python
def is_isomorphic_short(s, t):
    # Số lượng cặp (s[i], t[i]) duy nhất phải bằng số lượng ký tự duy nhất trong s và t
    return len(set(s)) == len(set(t)) == len(set(zip(s, t)))

s, t = map(str, input("Nhập 2 từ có độ dài bằng nhau: ").split())
if is_isomorphic_short(s, t):
    print("Hai từ này đẳng cấu")
else:
    print("Hai từ này không đẳng cấu")
```

#### Bài 12
#### Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và 0 <= |r| < |divisor|.

```python
import math
def divide_with_remainder(dividend, divisor):
    if divisor == 0:
        return "Số chia không thể bằng 0"
    q, r = divmod(dividend, divisor)
    return q, r
a, b = map(int, input("Nhập 2 số a và b (mỗi số cách nhau 1 dấu cách): ").split())
print(divide_with_remainder(a, b))
```

#### Bài 15
#### Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
def is_right_triangle(a, b, c):
    if a + b > c and b + c > a and a + c > b and a > 0 and b > 0 and c > 0:
        if (a ** 2) + (b ** 2) == (c ** 2) or (c ** 2) + (b ** 2) == (a ** 2) or (a ** 2) + (c ** 2) == (b ** 2):
            return True
        else:
            return False
    return False

a, b, c = map(int, input("Nhập 3 số a, b, c: ").split())
print(is_right_triangle(a, b, c))
```
