#### Bài 2
####  Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2 chữ số phần thập phân. 


```python
a = float(input("Nhập chiều rộng hình chữ nhật: "))
b = float(input("Nhập chiều dài hình chữ nhật: "))
if a <=0 or b <= 0:
    print("Chỉ nhập số dương")
elif a > b:
    print("Chiều dài phải lớn hơn chiều rộng")
else:
    s = (a * b) - (((a / 2) ** 2 ) * 3.14)
    print(f"Diện tích phần còn lại để trồng cây: {s:.2f}")
```

#### Bài 3
#### Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó.

```python
c = input("Nhập 1 kí tự: ")
if len(c) == 1:
    if c.isupper():
        print(c.lower())
    elif c.islower():
        print(c.upper())
    else:
        print(f"{c} không phải chữ cái")
else:
    print("Chỉ nhập 1 kí tự")
```

#### Bài 4
#### Viết chương trình nhập vào một “kí tự”, Kiểm tra xem c có phải là kí tự chữ cái không (hoa hoặc thường). Nếu đúng thì in ra “{c} là kí tu alphabet”, nếu sai in ra “{c} không phải là kí tự alphabet.

```python
c = input("Nhập vào 1 kí tự: ")
if len(c) == 1:
    if 'a' <= c <= 'z' or 'A' <= c <= 'Z':
        print(f"{c} là kí tự alphabet")
    else:
        print(f"{c} không phải kí tự alphabet")
else:
    print("Chỉ nhập 1 kí tự")
```

#### Bài 5
#### Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương ứng của A. Xét TH khi c = ‘A’ là đặc biệt. 

```python
c = input("Nhập vào một chữ cái hoa (A-Z): ")
if 'A' <= c <= 'Z':
    char_lower = c.lower()
    if c == 'A':
        result = 'z'
    else:
        result = chr(ord(char_lower) - 1)     
    print(f"Chữ cái thường liền trước của '{char_lower}' là: {result}")
else:
    print("Vui lòng nhập đúng một chữ cái IN HOA.")
```

#### Bài 6
#### Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”. 

```python
data = input("Nhập vào 3 số: ").split()
if len(data) != 3:
    print("Nhập đúng 3 số")
else:
    a, b, c = map(float, data)
    p = (a + b + c) / 2
    if a + b > c and b + c > a and a + c > b:
        s = (p * (p - a) * (p - b) * (p - c)) ** 0.5
        print(f"Diện tích tam giác: {s:.1f}")
    else:
        print("Không phải 3 cạnh 1 tam giác")
```

#### Nhập điểm (0-10), phân loại học lực theo quy tắc: >=8: Giỏi; >=6.5 và <8: Khá; >=5 và <6.5: Trung bình; <5: Yếu

```python
a = float(input("Nhập điểm: "))
if a < 0 or a > 10:
    print("Điểm phải nằm trong khoảng từ 0 đến 10")
else:
    if a >= 8:
        print("Học lực giỏi")
    elif 6.5 <= a < 8:
        print("Học lực khá")
    elif 5 <= a < 6.5:
        print("Học lực trung bình")
    else:
        print("Học lực yếu")
```

#### Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số tương ứng (0 -> "Không", 1 -> "Một", ...) 

```python
n = int(input("Nhập số: "))
match n:
    case 0: print("Không")
    case 1: print("Một")
    case 2: print("Hai")
    case 3: print("Ba")
    case 4: print("Bốn")
    case 5: print("Năm")
    case 6: print("Sáu")
    case 7: print("Bảy")
    case 8: print("Tám")
    case 9: print("Chín")
    case _: print("Số nằm ngoài phạm vi 0-9")
```

#### Kiểm tra điểm thi xem có qua môn hay phải học lại: Nhập điểm thi, nếu điểm >= 4 thì in "Qua môn", ngược lại in "Học lại". 

```python
a = float(input("Nhập điểm môn học: "))
if 0 <= a <= 10:
    if a >= 4:
        print("Qua môn")
    else:
        print("Học lại")
else:
    print("Điểm phải nằm trong khoảng từ 0 đến 10")
```
