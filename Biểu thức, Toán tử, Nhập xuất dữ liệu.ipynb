#### Bài 3
#### Viết chương trình nhận vào hai số nguyên a, b từ bàn phím. In ra tổng, hiệu, tích, phần nguyên, phần dư và kết quả chia thực của phép chia a cho b (lấy đến 2 chữ số phần thập phân)

```python
a = int(input("Nhập số nguyên a: "))
b = int(input("Nhập số nguyên b: "))
if b == 0:
    print("Không thể thực hiện phép chia vì số b bằng 0.")
else:
    tong = a + b
    hieu = a - b
    tich = a * b
    phan_nguyen = a // b
    phan_du = a % b
    chia_thuc = a / b
    print(f"Tổng: {tong}")
    print(f"Hiệu: {hieu}")
    print(f"Tích: {tich}")
    print(f"Phần nguyên: {phan_nguyen}")
    print(f"Phần dư: {phan_du}")
    print(f"Chia thực: {chia_thuc:.2f}")
```

#### Bài 4
#### Viết chương trình python để tính điểm trung bình môn học của một học sinh trong lớp. Có tổng cộng 6 đầu điểm: a1, b1, c1, a2, b2, a3. Trong đó:
#### • Các điểm a1, b1, c1 được nhân hệ số 1,
#### • Các điểm a2, b2 được nhân hệ số 2,
#### • Điểm a3 được nhân hệ số 3.
#### Công thức tính điểm trung bình là: TB=((a1+b1+c1)+(a2+b2)×2+a3×3)/10
#### Đầu vào: Trong bài tập này, dữ liệu vào chỉ gồm duy nhất một dòng bao gồm 6 số thể hiện cho 6 đầu điểm a1, b1, c1, a2, b2, a3 theo đúng thứ tự đó. Trong đó, hai đầu điểm liên tiếp nhau sẽ được phân tách nhau bởi duy nhất 1 dấu cách
#### Đầu ra: Đầu ra là giá trị điểm trung bình môn học của 6 môn học, lấy đến 1 chữ số phần thập phân

```python
data = input("Nhập 6 đầu điểm (cách nhau bởi dấu cách): ").split()   
if len(data) != 6:
    print("Vui lòng nhập chính xác 6 đầu điểm!")
else:
    a1, b1, c1, a2, b2, a3 = map(float, data)
    diem_tb = ((a1 + b1 + c1) + (a2 + b2) * 2 + a3 * 3) / 10
    print(f"Điểm trung bình của học sinh là: {diem_tb:.1f}")
```

#### Bài 6
#### Hãy nhập vào một kí tự chữ cái thường ‘a’…’z’, in ra mã Unicode của nó sau đó chuyển nó thành kí tự hoa tương ứng, in kí tự hoa này ra màn hình.

```python
char = input("Nhập một kí tự chữ cái thường (a-z): ")
if len(char) == 1 and 'a' <= char <= 'z':
    ma_unicode = ord(char)
    print(f"Mã Unicode của '{char}' là: {ma_unicode}")
    chu_hoa = char.upper()
    print(f"Kí tự hoa tương ứng là: {chu_hoa}")
else:
    print("Vui lòng chỉ nhập duy nhất một kí tự chữ cái thường từ a đến z!")
```

#### Bài 8
#### Fahrenheit, hay độ F, là một thang nhiệt độ được đặt theo tên nhà vật lý người Đức Daniel Gabriel Fahrenheit (1686–1736). Dựa vào nhiệt độ Celcius được nhập từ bàn phím, hãy chuyển sang độ F theo công thức: F = 9/5*C + 32 và in ra màn hình kết quả. Kết quả làm tròn đến hai chữ số thập phân đằng sau dấu phẩy.

```python
do_celcius = float(input("Nhập nhiệt độ theo đơn vị độ C: "))
do_fahrenheit = ((9 / 5) * do_celcius) + 32
print(f"Nhiệt độ theo đơn vị độ F: {do_fahrenheit:.2f}")
```

Bởi Amazon không vận chuyển trực tiếp về Việt Nam, Đạt phải trả 10 USD cho phí vận chuyển từ một nước trung gian. Khi vận chuyển về Việt Nam, Đạt được thông báo sẽ phải trả thêm 30% thuế xuất nhập cảnh và 10% thuế VAT cho chiếc đồng hồ nói trên. Viết chương trình tính tổng số tiền mà Đạt phải trả để mua được chiếc đồng hồ trên (USD). Kết quả làm tròn đến hai chữ số thập phân.


```python
x = float(input("nhập giá tiền chiếc đồng hồ: "))
tien = (x + (x * (3 / 10))) + (x + (x * (1 / 10))) + 10
print(f"Số tiền Đạt phải trả: {tien:.2f} USD")
```

#### Đầu ra: In ra màn hình một dòng duy nhất số giờ và phút được chuyển sang đơn vị giây.

#### Bài 15
#### Viết chương trình nhận vào một số nguyên dương n và trả về "số sao" thứ n.
#### Một "số sao" là một số hình học trung tâm đại diện cho một hình lục giác có tâm (ngôi sao sáu cánh), giống như hình dạng được sử dụng trong trò chơi cờ Trung Quốc.

```python
n = int(input("Nhập số nguyên dương: "))
if n <= 0:
    print("Chỉ nhập số nguyên dương")
else:
    so_sao = (6 * n * (n - 1)) + 1
    print(f"Số sao thứ n: {so_sao}")
```
