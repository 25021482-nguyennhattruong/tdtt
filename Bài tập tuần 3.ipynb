#### Bài 1
#### Nhập vào một số nguyên dương có 3 chữ số. Tìm và in ra màn hình “SỐ NGUYÊN” đảo ngược của n. Không dùng kiểu chuỗi. Chú ý: kiểm tra chỉ xử lý các số nguyên dương có đúng 3 chữ số. Các số khác không xử lý.

```python
n = int(input("Nhập một số nguyên dương có 3 chữ số: "))
if 100 <= n <= 999:
    hang_tram = n // 100
    hang_chuc = (n // 10) % 10
    hang_don_vi = n % 10
    so_dao_nguoc = (hang_don_vi * 100) + (hang_chuc * 10) + hang_tram    
    print(f"Số nguyên đảo ngược của {n} là: {so_dao_nguoc}")
else:
    print("Số nhập vào không phải là số có 3 chữ số.")
```

#### Bài 3
#### Kiểm tra xem một số có phải là lũy thừa của 2 hay không
#### Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không. 

```python
n = int(input("Nhập 1 số nguyên dương: "))
if n > 0:
    if n & (n - 1) == 0:
        print(f"{n} là lũy thừa của 2")
    else:
        print(f"{n} không phải lũy thừa của 2")
else:
    print("Vui lòng nhập số nguyên dương")
```

```python
a = input("Nhập vào chuỗi a: ")
b = input("Nhập vào chuỗi b: ")
if a.isalpha() and b.isalpha():
    if a.islower() and b.islower():
        if len(a) > len(b):
            print("True")
        else:
            print("False")
    else:
        print("Chỉ nhập chữ cái thường")
else:
    print("Chỉ nhập chữ cái thường")
```

#### Bài 11
#### Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm tra 3 số đã nhập có tạo thành một tam giác hợp lệ hay không. Nếu không in ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay "Tam giác thường".

```python
a, b, c = map(int, input("Nhập vào 3 số nguyên dương (mỗi số cách nhau 1 dấu cách): ").split())
if a > 0 and b > 0 and c > 0:
    if a + b > c and b + c > a and a + c > b:
        if a == b == c:
            print("Tam giác đều")
        elif a == b or b == c or c == a:
            if (a ** 2) + (b ** 2) == (c ** 2) or (b ** 2) + (c ** 2) == (a ** 2) or (a ** 2) + (c ** 2) == (b ** 2):
                print("Tam giác vuông cân")
            else:
                print("Tam giác cân")
        elif (a ** 2) + (b ** 2) == (c ** 2) or (b ** 2) + (c ** 2) == (a ** 2) or (a ** 2) + (c ** 2) == (b ** 2):
            print("Tam giác vuông")
        else:
            print("Tam giác thường")
    else:
        print("Không phải tam giác")
else:
    print("Chỉ nhập số nguyên dương")
```

#### Bài 17
#### Cho 4 số a, b, c, d. Hãy kiểm tra xem 4 số này có thể theo thứ tự tạo thành 1 cấp số nhân với công bội nguyên theo đúng thứ tự a, b, c, d hay không?

```python
a, b, c, d = map(int, input("Nhập 4 số a, b, c, d cách nhau bởi dấu cách: ").split())
if a == 0:
    if b == 0 and c == 0 and d == 0:
        print("Yes")
    else:
        print("No")
else:
    if b % a == 0:
        q = b / a
        if c == b * q and d == c * q:
            print("Yes")
        else:
            print("No")
    else:
        print("No")
```
