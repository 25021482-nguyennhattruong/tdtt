#### Bài 4
#### Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key lặp lại, thêm vào list của key đó

```python
# Nhập chuỗi và tách thành các cặp
data = input().split()
res = {}

for item in data:
    # Tách key và value từ mỗi cặp
    key, value = item.split(':')
    
    # Nếu key chưa có trong dict, tạo một list trống
    if key not in res:
        res[key] = []
    
    # Thêm value vào list của key đó
    res[key].append(value)

print(res)
```

#### Bài 5
#### Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 'zeros': ...} (đếm số >0, <0, =0)

```python
n = list(map(int, input().split()))
res = {
    'positives':0,
    'negatives':0,
    'zeros':0
}
for i in n:
    if i > 0:
        res['positives'] += 1
    elif i < 0:
        res['negatives'] += 1
    else:
        res['zeros'] += 1
print(res)
```

#### Bài 10
#### Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.

```python
try:
    numbers = list(map(int, input("Nhập danh sách các số nguyên (cách nhau bởi dấu cách): ").split()))
    k = int(input("Nhập số nguyên dương k: "))
except ValueError:
    print("Vui lòng chỉ nhập số nguyên!")
    exit()

# 2. Tìm các cặp số thỏa mãn điều kiện
pairs = []
n = len(numbers)

# Duyệt qua từng cặp (i, j) với i < j để đảm bảo thứ tự xuất hiện
for i in range(n):
    for j in range(i + 1, n):
        if numbers[i] + numbers[j] == k:
            # Tạo tuple (số thứ nhất, số thứ hai)
            pairs.append((numbers[i], numbers[j]))

# 3. Loại bỏ các cặp trùng lặp (nếu cần - tùy vào yêu cầu "khác nhau")
unique_pairs = list(set(pairs))

# 4. Sắp xếp list các tuple theo phần tử đầu tiên
unique_pairs.sort(key=lambda x: x[0])
print(unique_pairs)
```

#### Bài 12
#### Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ nhất trong số đó.

```python
n = list(map(int, input().split()))
min = n[0]
dem = 1
for i in n:
    if n.count(i) >= dem:
        if min > i:
            min = i
            dem = n.count(i)
print(min)
```

#### Bài 13
#### Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo một dictionary mới bằng cách đảo ngược key và value của dictionary ban đầu. In ra dictionary mới.

```python
def nhap(dict):
    d = {}
    n = int(input("Nhập số lượng phần tử: "))
    if n == 0:
        return d
    else:
        for _ in range(n):
            key = input("Nhập key: ")
            value = int(input("Nhập giá trị cho key: "))
            d[key] = value
        return d
    
old_dict = nhap("Nhập dictionary: ")
new_dict = {value: key for key, value in old_dict.items()}
print(new_dict)
```

#### Bài 15
#### Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp keyvalue từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary mới.

```python
def nhap(dict):
    d = {}
    n = int(input("Nhập số lượng phần tử: "))
    if n == 0:
        return d
    else:
        for _ in range(n):
            key = input("Nhập key: ")
            value = int(input("Nhập giá trị cho key: "))
            d[key] = value
        return d
    
dict = nhap("Nhập dictionary: ")
k = int(input())
d = {}
for key, value in dict.items():
    if value > k:
        d[key] = value
print(d)
```
