```python
def so_chinh_phuong(n):
    for i in range(1, int(n ** 0.5) + 1):
        if i ** 2 == n:
            return True
    return False

x = int(input())
print(bool(so_chinh_phuong(x)))
```

#### Bài 20
####  Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.

```python
def detect_number_type(s):
    # Thử chuyển sang kiểu int đầu tiên
    try:
        # Nếu s chứa dấu chấm, int() sẽ lỗi ngay, ta sẽ chuyển sang float
        if '.' in s:
            float(s)
            return 'float'
        
        int(s)
        return 'int'
    except ValueError:
        # Nếu không phải int, thử chuyển sang float
        try:
            float(s)
            return 'float'
        except ValueError:
            # Nếu cả hai đều lỗi thì không phải số
            return 'neither'

n = input()
print(detect_number_type(n))
```

#### Bài 24
#### Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.

```python
def are_conllinear(x1, y1, x2, y2, x3, y3):
    det = (x1 * y2) + (y1 * x3) + (y3 * x2) - (x1 * y3) - (y2 * x3) - (x2 * y1)
    if det == 0:
        return "3 điểm thẳng hàng"
    else:
        return "3 điểm không thẳng hàng"
    
x1, y1, x2, y2, x3, y3 = map(float, input().split())
print(are_conllinear(x1, y1, x2, y2, x3, y3))
```

#### Bài 25
#### Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3), (0,3)], p = (2,1)

```python
def point_in_quadrilateral(quad, p):
    """
    Kiểm tra điểm p(x, y) có nằm trong hoặc trên biên của tứ giác quad hay không.
    quad: Danh sách 4 tuple tọa độ [(x1, y1), (x2, y2), (x3, y3), (x4, y4)]
    p: Tuple tọa độ điểm cần kiểm tra (x, y)
    """
    def cross_product(a, b, c):
        # Tính tích có hướng của vector AB và AC
        # (Bx - Ax) * (Cy - Ay) - (By - Ay) * (Cx - Ax)
        return (b[0] - a[0]) * (c[1] - a[1]) - (b[1] - a[1]) * (c[0] - a[0])

    n = len(quad)
    signs = []
    
    for i in range(n):
        # Lấy hai đỉnh liên tiếp tạo thành một cạnh
        p1 = quad[i]
        p2 = quad[(i + 1) % n] # Đỉnh kế tiếp (nối đỉnh cuối về đỉnh đầu)
        
        cp = cross_product(p1, p2, p)
        
        # Nếu tích có hướng = 0, điểm p nằm trên đường thẳng chứa cạnh
        # Ta cần lưu dấu để kiểm tra tính đồng nhất
        if cp > 0:
            signs.append(1)
        elif cp < 0:
            signs.append(-1)
        else:
            # Điểm nằm trên biên hoặc đường thẳng kéo dài của cạnh
            # Trong ngữ cảnh bài này, ta coi như thỏa mãn dấu của các cạnh khác
            pass

    # Nếu tất cả các tích có hướng không bằng 0 đều cùng dấu (toàn dương hoặc toàn âm) thì điểm p nằm bên trong hoặc trên biên.
    return all(s > 0 for s in signs) or all(s < 0 for s in signs) or not signs

# Ví dụ sử dụng:
quad = [(0, 0), (4, 0), (4, 3), (0, 3)]
p = (2, 1)

if point_in_quadrilateral(quad, p):
    print(f"Điểm {p} nằm TRONG hoặc TRÊN BIÊN tứ giác.")
else:
    print(f"Điểm {p} nằm NGOÀI tứ giác.")
```

#### Bài 28
#### Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math

def tinh_goc_abc(A, B, C):
    # Tọa độ các điểm: A=(x1, y1), B=(x2, y2), C=(x3, y3)
    # Tính vector BA và BC
    ba = (A[0] - B[0], A[1] - B[1])
    bc = (C[0] - B[0], C[1] - B[1])

    # Tính tích vô hướng (dot product)
    dot_product = ba[0] * bc[0] + ba[1] * bc[1]

    # Tính độ dài (magnitude) của các vector
    mag_ba = math.sqrt(ba[0]**2 + ba[1]**2)
    mag_bc = math.sqrt(bc[0]**2 + bc[1]**2)

    # Tránh lỗi chia cho 0 nếu các điểm trùng nhau
    if mag_ba == 0 or mag_bc == 0:
        return 0

    # Tính cosin và giới hạn trong khoảng [-1, 1] để tránh lỗi làm tròn
    cosine_angle = dot_product / (mag_ba * mag_bc)
    cosine_angle = max(-1, min(1, cosine_angle))

    # Tính góc (radian) sau đó chuyển sang độ
    angle_rad = math.acos(cosine_angle)
    angle_deg = math.degrees(angle_rad)

    return angle_deg

# Ví dụ sử dụng:
A = (0, 1)
B = (0, 0)
C = (1, 0)

goc = tinh_goc_abc(A, B, C)
print(f"Góc ABC là: {goc:.2f}°")
```

#### Viết hàm tính hệ số nhị thức C(n, k).

```python
import math

def tinh_he_so_nhi_thuc(n, k):
    # Kiểm tra điều kiện thực tế của tổ hợp
    if k < 0 or k > n:
        return 0
    
    # Sử dụng hàm math.comb có sẵn trong Python 3.8+ 
    # Đây là cách tối ưu nhất để tránh tràn số và sai số
    return math.comb(n, k)

# Ví dụ sử dụng
n = int(input("Nhập n: "))
k = int(input("Nhập k: "))

ket_qua = tinh_he_so_nhi_thuc(n, k)
print(f"Hệ số nhị thức C({n}, {k}) là: {ket_qua}")
```

#### Bài 33
#### Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa, giá km, phí chờ, phụ phí ban đêm và giảm giá.

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    # 1. Định nghĩa giá theo loại xe
    if car_type == 4:
        price_per_km = 15000
    elif car_type == 7:
        price_per_km = 17000
    else:
        return "Loại xe không hợp lệ (chỉ chọn 4 hoặc 7)"

    # 2. Tính giá theo quãng đường
    opening_fare = 12000
    if km <= 1:
        fare = opening_fare
    else:
        fare = opening_fare + (km - 1) * price_per_km

    # 3. Tính phí chờ (miễn phí 5 phút đầu)
    if wait_minutes > 5:
        wait_fee = (wait_minutes - 5) * 1000
    else:
        wait_fee = 0
    
    total_fare = fare + wait_fee

    # 4. Áp dụng giảm giá đường dài (> 30km giảm 10%)
    if km > 30:
        total_fare *= 0.9

    # 5. Phụ phí ban đêm (từ 22h đêm đến 5h sáng)
    # Nếu giờ bắt đầu nằm trong khoảng này, tăng 20%
    if start_hour >= 22 or start_hour < 5:
        total_fare *= 1.2

    return round(total_fare)

# --- Ví dụ sử dụng ---
car = 4             # Xe 4 chỗ
hour = 23           # Đi lúc 11 giờ đêm
distance = 35       # Đi 35 km
waiting = 10        # Chờ 10 phút

final_price = smart_taxi_fare(car, hour, distance, waiting)
print(f"Tổng cước phí taxi: {final_price:,} VNĐ")
```

#### Bài 34
#### Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau. Nếu n ≤ 0 trả về 0.

```python
def check_unique_digits(x):
    """Kiểm tra nếu các chữ số của x đôi một khác nhau."""
    s = str(x)
    # Tập hợp (set) sẽ loại bỏ các phần tử trùng lặp
    # Nếu độ dài của set bằng độ dài của chuỗi ban đầu, nghĩa là các chữ số là duy nhất
    return len(set(s)) == len(s)

def sum_1_to_n(n):
    """Tính tổng các số từ 1 đến n có các chữ số đôi một khác nhau."""
    if n <= 0:
        return 0
    
    tong = 0
    for x in range(1, n + 1):
        if check_unique_digits(x):
            tong += x
            
    return tong

# --- Kiểm tra chương trình ---
n_input = int(input())
ket_qua = sum_1_to_n(n_input)
print(f"Tổng các số từ 1 đến {n_input} có chữ số đôi một khác nhau là: {ket_qua}")
# Ví dụ n=20: Các số thỏa mãn là 1->10 và 12->20 (loại số 11)
```
