#### Bài 1
#### Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để tính tống các số nguyên từ 1 đến n

#### Bài 2
#### VIết chương trình nhập vào một số tới khi được một số nguyên dương thì thôi. Kiểm tra số này có phải là số nguyên tố hay không? 

```python
def so_nguyen_to(n):
    if n <= 1:
        return False
    else:
        for i in range(2, int(n ** 0.5) + 1):
            if n % i == 0:
                return False
        return True
        
while True:
    try:
        n = int(input("Nhập vào một số nguyên dương n: "))
        if n > 0:
            break
        print("Vui lòng nhập số lớn hơn 0!")
    except ValueError:
        print("Đó không phải là số nguyên. Thử lại nhé!")

if so_nguyen_to(n):
    print(f"{n} là số nguyên tố")
else:
    print(f"{n} không phải số nguyên tố")
```

#### Bài 6
#### Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn [a, b]. Mỗi dòng là một bộ test (a b)

```python
def so_nguyen_to(n):
    if n <= 1:
        return False
    else:
        for i in range(2, int(n ** 0.5) + 1):
            if n % i == 0:
                return False
        return True

a, b = map(int, input("Nhập 2 số nguyên dương (mỗi số cách nhau 1 dấu cách): ").split())
if 0 < a <= b:
    tong = sum(i for i in range(a, b + 1) if so_nguyen_to(i))
    print(tong)
```

#### Bài 7
#### Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n

```python
def ưsntln(n):
  i = 2
  while i * i <= n:
    if n % i != 0:
      i += 1
    else:
      n //= i
  return n

n = int(input("Nhập số nguyên dương n: "))
if n >= 2:
  print(f"Ước số nguyên lớn nhất của {n} là {ưsntln(n)}")
else:
  print("Vui lòng nhập số nguyên dương lớn hơn hoặc bằng 2")
```

#### Bài 8
#### Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số đối xứng (palindrome). In ra số bước và giá trị palindrome thu được.

```python
def palindrome(n):
  return str(n) == str(n)[::-1]
def reverse(n):
  return int(str(n)[::-1])

n = int(input("Nhập số nguyên dương n: "))
if n > 0:
  buoc = 0
  while not palindrome(n):
    a = reverse(n)
    n += a
    buoc += 1
  print(f"Số bước: {buoc}")
  print(f"Giá trị palidrome thu được: {n}")
else:
  print("Vui lòng nhập số nguyên dương")
```

#### Bài 9
#### Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu cách). Nếu không có, in ra (no number)

```python
def has_unique_digits(num):
    # Chuyển số thành chuỗi để kiểm tra từng ký tự
    s = str(num)
    # Nếu độ dài chuỗi bằng kích thước của tập hợp các chữ số thì các chữ số là duy nhất
    return len(s) == len(set(s))

def solve():
    try:
        n = int(input("Nhập n: "))
        if n < 0:
            print("no number")
            return
    except ValueError:
        return

    result = []
    i = 1
    while True:
        square = i * i
        if square > n:
            break
        
        if has_unique_digits(square):
            result.append(str(square))
        i += 1

    if result:
        print(" ".join(result))
    else:
        print("(no number)")

solve()
```

#### Bài 10
#### Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số nhỏ nhất.

```python
def do_dai_collatz(n):
  do_dai = 1
  while n > 1:
    if n % 2 == 0:
      n = n // 2
    else:
      n = n * 3 + 1
    do_dai += 1
  return do_dai

n = int(input("Nhập số nguyên dương n: "))
if n <= 0:
    print("Vui lòng nhấp số nguyên dương")
else:
    so = 0
    max = 0
    for i in range(1, n + 1):
        if max < do_dai_collatz(i):
            max = do_dai_collatz(i)
            so = i
    print(f"Số có độ dài Collatz lớn nhất: {so}")
    print(f"Độ dài dãy Collatz của {so}: {max}")
```

#### Bài 13
#### Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật: Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại.
#### Ví dụ 220 và 284 là một cặp số thân thiết
#### 220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55, 110. Tổng của 11 ước này cho chúng ta 284
#### 284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng vừa đúng bằng 220
#### Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không. Nếu có, in ra màn hình "true", ngược lại, in ra "false" 

```python
a, b = map(int, input("Nhập 2 số nguyên dương (mỗi số cách nhau 1 dấu cách): ").split())
def uoc_so(n):
  tong = 0
  for i in range(1, n):
    if n % i == 0:
      tong += i
  return tong

if a == uoc_so(b) and b == uoc_so(a):
  print("true")
else:
  print("false")
```

#### Bài 14
#### Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình tìm ước chung lớn nhất của hai số m và n. 

```python
m, n = map(int, input("Nhập 2 số nguyên dương (mỗi số cách nhau 1 dấu cách): ").split())
def ucln(m, n):
  while n != 0:
    m, n = n, m % n
  return m
print(f"Ước chung lón nhất của {m} và {n}: {ucln(m, n)}")
```
