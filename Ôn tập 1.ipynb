#### Bài 1
#### Yêu cầu. Cước phí Taxi Thông Minh.
#### Viết một hàm duy nhất smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính và trả về total (tổng tiền phải trả, kiểu số nguyên, đơn vị VND).
#### Các quy tắc cần cài đặt:
#### 1. Giá mở cửa (0–0.8 km):
#### - Xe 4 chỗ: 11,000 VND
#### - Xe 7 chỗ: 12,000 VND
#### 2. Giá theo km sau 0.8 km:
#### - Xe 4 chỗ: từ 0.8→20 km tính 13,000 VND/km; trên 20 km tính 10,500 VND/km
#### - Xe 7 chỗ: từ 0.8→20 km tính 14,000 VND/km; trên 20 km tính 12,000 VND/km
#### 3. Phí chờ: 5 phút đầu miễn phí; sau đó, mỗi 5 phút (làm tròn lên theo từng block 5 phút) tính 2,000 VND.
#### 4. Phụ thu ban đêm: nếu start_hour thuộc [22,23,0,1,2,3,4,5], cộng +20% trên tiền tạm tính (mở cửa + quãng đường + chờ).
#### 5. Điều kiện đặc biệt:
#### - Nếu km > 30 và không phải ban đêm, áp dụng giảm 10% trên tổng tiền.
#### - Nếu tổng tiền cuối cùng < 50,000 nhưng km > 5, đặt tổng tiền bằng 50,000 VND (mức tối thiểu). 

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    if car_type == 4:
        if km < 0.8:
            gia = 11000
        elif 0.8 <= km < 20:
            gia = 11000 + 13000 * (km - 0.8)
        else:
            gia = 11000 + 13000 * (km - 0.8) + 10500 (km - 20)
    elif car_type == 7:
        if km < 0.8:
            gia = 12000
        elif 0.8 <= km < 20:
            gia = 11000 + 14000 * (km - 0.8)
        else:
            gia = 11000 + 14000 * (km - 0.8) + 12000 (km - 20)
    else:
        return "Loại xe không hợp lệ"
    
    if wait_minutes <= 5:
        phi_cho = 0
    else:
        phi_cho = int(wait_minutes / 5) * 2000
    tong = gia + phi_cho
    if start_hour >= 22 or start_hour <= 5:
        tong = tong + tong * (20/100)
    if km > 30 and start_hour >= 22 or start_hour <= 5:
        tong = tong - tong * (30/100)
    if tong < 50000 and km > 5:
        tong = 50000
    return int(tong)

car_type = int(input())
start_hour = float(input())
km = float(input())
wait_minutes = int(input())
print(smart_taxi_fare(car_type, start_hour, km, wait_minutes))
```

#### Bài 2
#### Cho số nguyên x, trả về True nếu tồn tại số nguyên n ≥ 1 sao cho x = 1 + 2 + ... + n, ngược lại trả về False. Khuyến khích dùng vòng lặp đơn giản để cộng dồn tổng.

```python
def so_sanh(x):
    tong = 0
    n = 0
    while tong + (n + 1) <= x:
        n += 1
        tong += n
    return tong == x

x = int(input())
print(so_sanh(x))
```

#### Bài 4
#### Yêu cầu: Cho các hệ số thực a, b, c, trả về tất cả các nghiệm thực của phương trình bậc hai axx + bx + c = 0 trong một danh sách được sắp xếp tăng dần.
#### Quy tắc:
#### - Nếu a == 0 và b != 0: trả về nghiệm duy nhất [-c / b].
#### - Nếu a == 0 và b == 0: trả về [] (không có nghiệm hữu hạn duy nhất nếu c == 0, hoặc vô nghiệm nếu c != 0).
#### - Với a != 0, dùng biệt thức d = bb - 4ac:
#### - d < 0 → []
#### - d == 0 → [ -b / (2a) ]
#### - d > 0 → hai nghiệm [r1, r2] với r1 <= r2.

```python
def nghiem_thuc(a, b, c):
    tap_nghiem = []
    if a == 0:
        if b != 0:
            tap_nghiem.append((-c) / b)
    else:
        d = b ** 2 - (4 * a * c)
        if  d == 0:
            tap_nghiem.append(-b / (2 * a))
        elif d > 0:
            tap_nghiem.append((-b + (d ** 0.5)) / (2 * a))
            tap_nghiem.append((-b - (d ** 0.5)) / (2 * a))
    return sorted(tap_nghiem)

a, b, c = map(float, input().split())
print(nghiem_thuc(a, b, c))
```

#### Bài 5
#### Cho ba số a, b, c, trả về chúng dưới dạng danh sách theo thứ tự không giảm.

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

a = list(map(float, input().split()))
print(selection_sort(a))
```

#### Bài 7
#### Yêu cầu: Cho ba số a, b, c, trả về giá trị ở giữa khi sắp xếp ba số theo thứ tự tăng dần (giá trị lớn thứ hai). Định nghĩa này vẫn đúng khi có phần tử trùng nhau (ví dụ: 2, 2, 5 → giá trị giữa là 2).

```python
def gia_tri_o_giua(a, b, c):
    list = [a, b, c]
    n = len(list)
    for i in range(n):
        min = i
        for j in range(i + 1, n):
            if list[j] < list[min]:
                min = j
        list[i], list[min] = list[min], list[i]
    return list[1]

a, b, c = map(float, input().split())
print(gia_tri_o_giua(a, b, c))
```

#### Bài 16
#### Viết hàm trả về góc ∠ABC theo đơn vị độ khi biết tọa độ các điểm A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math

def calculate_angle_abc(x1, y1, x2, y2, x3, y3):
    # 1. Tính tọa độ các vectơ BA và BC (B là gốc)
    ba_x, ba_y = x1 - x2, y1 - y2
    bc_x, bc_y = x3 - x2, y3 - y2

    # 2. Tính tích vô hướng (dot product) của hai vectơ
    dot_product = ba_x * bc_x + ba_y * bc_y

    # 3. Tính độ dài (magnitude) của mỗi vectơ
    mag_ba = math.sqrt(ba_x**2 + ba_y**2)
    mag_bc = math.sqrt(bc_x**2 + bc_y**2)

    # Kiểm tra tránh lỗi chia cho 0 nếu điểm A hoặc C trùng với B
    if mag_ba == 0 or mag_bc == 0:
        raise ValueError("Các điểm không được trùng nhau để tạo thành góc.")

    # 4. Tính Cosine của góc (giới hạn trong khoảng [-1, 1] để tránh lỗi làm tròn)
    cosine_angle = max(-1, min(1, dot_product / (mag_ba * mag_bc)))

    # 5. Tính góc theo radian và chuyển sang độ
    angle_rad = math.acos(cosine_angle)
    angle_deg = math.degrees(angle_rad)

    return angle_deg

x1, y1, x2, y2, x3, y3 = map(float, input().split())
print(calculate_angle_abc(x1, y1, x2, y2, x3, y3))
```

#### Bài 25
#### Yêu cầu: Viết hàm point_in_quadrilateral(quad, p) trả về True nếu điểm p nằm bên trong hoặc trên biên của tứ giác quad, và False nếu ngược lại.
#### - Đầu vào:
#### - quad: danh sách hoặc bộ gồm đúng 4 đỉnh [(x1,y1), (x2,y2), (x3,y3), (x4,y4)] theo thứ tự (tứ giác có thể lồi hoặc lõm, nhưng không tự cắt).
#### - p: một điểm (x, y).
#### - Đầu ra: giá trị boolean.
#### - Gợi ý: Cài đặt phép kiểm tra điểm trong đa giác tổng quát (ví dụ: ray casting), đồng thời tính cả trường hợp điểm nằm trên biên là ở trong. 


```python
def point_in_quadrilateral(quad, p):
  
    def cross_product(a, b, c):
        # Tính tích có hướng của vector AB và AC
        # (Bx - Ax) * (Cy - Ay) - (By - Ay) * (Cx - Ax)
        return (b[0] - a[0]) * (c[1] - a[1]) - (b[1] - a[1]) * (c[0] - a[0])

    n = len(quad)
    signs = []
    
    for i in range(n):
        # Lấy hai đỉnh liên tiếp tạo thành một cạnh
        p1 = quad[i]
        p2 = quad[(i + 1) % n] # Đỉnh kế tiếp (nối đỉnh cuối về đỉnh đầu)
        
        cp = cross_product(p1, p2, p)
        
        # Nếu tích có hướng = 0, điểm p nằm trên đường thẳng chứa cạnh
        # Ta cần lưu dấu để kiểm tra tính đồng nhất
        if cp > 0:
            signs.append(1)
        elif cp < 0:
            signs.append(-1)
        else:
            # Điểm nằm trên biên hoặc đường thẳng kéo dài của cạnh
            # Trong ngữ cảnh bài này, ta coi như thỏa mãn dấu của các cạnh khác
            pass

    # Nếu tất cả các tích có hướng không bằng 0 đều cùng dấu (toàn dương hoặc toàn âm) thì điểm p nằm bên trong hoặc trên biên.
    return all(s > 0 for s in signs) or all(s < 0 for s in signs) or not signs

# Ví dụ sử dụng:
quad = [(0, 0), (4, 0), (4, 3), (0, 3)]
p = (2, 1)

print(point_in_quadrilateral(quad, p))
```

#### Bài 26
#### Yêu cầu: Viết hàm are_collinear(x1, y1, x2, y2, x3, y3) trả về True nếu ba điểm thẳng hàng, ngược lại trả về False.
#### - Đầu vào: ba điểm x1, y1, x2, y2, x3, y3 biểu diễn tọa độ của ba điểm.
#### - Đầu ra: giá trị boolean.
#### - Gợi ý: Dùng phép kiểm tra diện tích (tích có hướng) và một sai số nhỏ khi so sánh số thực. Các điểm nằm trên cùng một đường thẳng được xem là thẳng hàng (kể cả khi có điểm trùng nhau).

```python
def are_collinear(x1, y1, x2, y2, x3, y3):
    det = x1 * (y2 - y3) + x2 * (y3 - y1) + x3 * (y1 - y2)
    return abs(det) < 1e-9

x1, y1, x2, y2, x3, y3 = map(float, input().split())
print(are_collinear(x1, y1, x2, y2, x3, y3))
```

#### Yêu cầu: Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình ax^2 + bx + c = 0.
#### - Đầu vào: các số thực a, b, c.
#### - Đầu ra: 0, 1, hoặc 2. Nếu phương trình suy biến thành 0 = 0 (tức a == 0, b == 0, c == 0), trả về float('inf').
#### - Quy tắc:
#### - Nếu a != 0, dùng biệt thức D = b^2 - 4ac.
#### - Nếu a == 0 và b != 0, đó là phương trình bậc nhất với đúng một nghiệm thực.
#### - Nếu a == 0 và b == 0 và c != 0, phương trình vô nghiệm.

```python
def quadratic_roots_count(a, b, c):
    if a == 0:
        if b == 0:
            if c == 0:
                return float('inf')
            else:
                return 0
        else:
            return 1
    else:
        deta = b ** 2 - 4 * a * c
        if deta > 0:
            return 2
        elif deta == 0:
            return 1
        else:
            return 0
        
a, b, c = map(float, input().split())
print(quadratic_roots_count(a, b, c))
```

#### Yêu cầu: Viết hàm sum_halves(s) nhận chuỗi s gồm đúng 6 chữ số và trả về tổng của số nguyên 3 chữ số đầu và số nguyên 3 chữ số cuối.
#### - Đầu vào: chuỗi s độ dài 6, chỉ gồm các chữ số 0-9.
#### - Đầu ra: tổng kiểu số nguyên.

#### - Nếu đầu vào không hợp lệ (sai độ dài hoặc chứa ký tự không phải chữ số), phát sinh ValueError.

```python
def sum_halves(s):
    if len(s) != 6 or not s.isdigit():
        raise ValueError
    else:
        a = s[:3]
        b = s[3:]
        return int(a) + int(b)
    
try:
    s = input()
    print(sum_halves(s))
except ValueError:
    print("Vui lòng nhập dãy có độ dài 6 ký tự và chỉ nhập chữ số")
```

#### Yêu cầu: Viết hàm detect_number_type(s) trả về 'int' nếu s biểu diễn một số nguyên, 'float' nếu biểu diễn một số thực dấu chấm động, ngược lại trả về 'neither'.
#### - Đầu vào: chuỗi s.
#### - Đầu ra: một trong các giá trị {'int', 'float', 'neither'}.
#### - Quy tắc:
#### - Cho phép có dấu + hoặc - ở đầu.
#### - Số nguyên phải chỉ gồm các chữ số và không có dấu chấm thập phân (ví dụ: '100').
#### - Số thực phải có dấu chấm và có chữ số ở cả hai phía của dấu chấm (ví dụ: '100.0', '-3.14').
#### - Các chuỗi như '.5', '5.', '1e3' được xem là 'neither' trong bài này.

```python
def detect_number_type(s):
    try:
        if '.' in s:
            float(s)
            return 'float'
        int(s)
        return 'int'
    except ValueError:
        try:
            float(s)
            return 'float'
        except ValueError:
            return 'neither'

n = input()
print(detect_number_type(n))
```

#### Bài 31
#### Yêu cầu: Viết hàm is_perfect_square(n) trả về True nếu n là số chính phương, ngược lại trả về False.
#### - Đầu vào: số nguyên n (có thể âm).
#### - Đầu ra: giá trị boolean.
#### - Lưu ý: Số âm không phải là số chính phương trong tập số nguyên.

```python
def is_perfect_square(n):
    if n < 0:
        return False
    for i in range(1, int(n ** 0.5) + 1):
        if n == i ** 2:
            return True
    return False
        
try:
    n = int(input())
    print(is_perfect_square(n))
except ValueError:
    print("Vui lòng nhập 1 số nguyên")
```

#### Yêu cầu: Viết hàm is_right_triangle(a, b, c) trả về True nếu a, b, c có thể là ba cạnh của một tam giác vuông, ngược lại trả về False.
#### - Đầu vào: các số dương a, b, c.
#### - Đầu ra: giá trị boolean.
#### - Lưu ý: Kiểm tra bất đẳng thức tam giác và định lý Pythagore. Dùng sai số nhỏ khi so sánh số thực.

```python
def is_right_triangle(a, b, c):
    if a + b > c or a + c > b or b + c > a:
        return (a ** 2) + (b ** 2) == (c ** 2) or (a ** 2) + (c ** 2) == (b ** 2) or (c ** 2) + (b ** 2) == (a ** 2)
    return False

a, b, c = map(float, input().split())
if a > 0 and b > 0 and c > 0:
    print(is_right_triangle(a, b, c))
else:
    print("Vui lòng chỉ nhập số dương")
```

#### Bài 33
#### Yêu cầu: Viết hàm compute(a, b, op) tính a (+|-||/) b tùy theo op.
#### - Đầu vào: các số a, b và ký tự op thuộc {'+', '-', '', '/'}.
#### - Đầu ra: kết quả số học.
#### - Yêu cầu thêm:
#### - Nếu op là '/' và b == 0, phát sinh ValueError.
#### - Nếu op không hợp lệ, phát sinh ValueError.

```python
def may_tinh(a, b, op):
    ket_qua = 0
    if op == '+':
        ket_qua = a + b
    elif op == '-':
        ket_qua = a - b
    elif op == '*':
        ket_qua = a * b
    elif op == '/':
        if b != 0:
            ket_qua = a / b
        else:
            raise ValueError("Không thể chia cho không")
    else:
        raise ValueError("Toán tử không hợp lệ")
    return round(ket_qua, 2)
try:
    a, b = map(float, input().split())
    op = input()
    print(may_tinh(a, b, op))
except ValueError as e:
    print(e)
```

#### Bài 35
#### Yêu cầu: Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) trong đó q là thương nguyên và r là số dư sao cho dividend = divisorq + r với 0 <= |r| < |divisor| (dùng ngữ nghĩa của Python với // và %).
#### - Đầu vào: số nguyên dividend, divisor (divisor khác 0).
#### - Đầu ra: bộ hai số nguyên (q, r).
#### - Ví dụ: divide_with_remainder(5, 3) -> (1, 2).

```python
import math
def divide_with_remainder(dividend, divisor):
    if divisor == 0:
        return "Số chia không thể bằng 0"
    q, r = divmod(dividend, divisor)
    return q, r
a, b = map(int, input().split())
print(divide_with_remainder(a, b))
```

#### Bài 37
#### Viết hàm trả về tam giác vuông cân cao n bằng ký tự ''. Dòng thứ i (đánh số từ 1) có i dấu '', các dòng nối với nhau bằng ký tự xuống dòng.

```python
def hinh(n):
    result = []
    for i in range(1, n + 1):
        result.append('"' * i)
    return "\n".join(result)

n = int(input())
print(hinh(n))
```

#### Bài 41
#### Viết hàm trả về ký tự đứng sau ký tự x n vị trí trong bảng chữ cái. Giữ nguyên chữ hoa/chữ thường và quay vòng trong phạm vi A..Z hoặc a..z.

```python
def shift_character(x, n):
    if 'A' <= x <= 'Z':
        return chr((ord(x) - ord('A') + n) % 26 + ord('A'))
    elif 'a' <= x <= 'z':
        return chr((ord(x) - ord('a') + n) % 26 + ord('a'))

x = input()
n = int(input())
print(shift_character(x, n))
```

```python
def dien_tich(a, b, c):
    if a > 0 and b > 0 and c > 0:
        if a + b > c or b + c > a or a + c > b:
            p = (a + b + c) / 2
            s = (p * (p - a) * (p - b) * (p - c)) ** 0.5
    return s

a, b, c = map(float, input().split())
print(dien_tich(a, b, c))
```

#### Bài 45
#### Viết hàm trả về số lượng số nguyên dương k sao cho 1 <= k < n và k > (n - 1) / 2. (Ví dụ: n=9 -> 4).

```python
def so_nguyen_duong(n):
    so = 0
    for k in range(1, n):
        if k > (n - 1) / 2:
            so += 1
    return so

n = int(input())
print(so_nguyen_duong(n))
```
