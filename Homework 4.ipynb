#### Bài 20
####  Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
n = int(input("Nhập số nguyên dương n: "))
if n <= 0:
    print("Vui lòng nhập số nguyên dương")
else:
    a = n
    while n % 2 == 0:
        n //= 2
    if n == 1:
        print(f"{a} có dạng 2^k")
    else:
        print(f"{a} không có dạng 2^k")
```

#### Bài 21
#### Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
n = int(input("Nhập số nguyên dương n: "))
if n > 0:
    tong = 0
    a = n
    while n > 0:
        chu_so = n % 10
        tong += chu_so
        n //= 10
    print(f"Tổng các chữ số trong {a}: {tong}")
else:
    print("Vui lòng nhập số nguyên dương")
```

#### Bài 22
#### Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input("Nhập số nguyên dương n: "))
if n <= 0:
    print("Vui lòng nhập số nguyên dương")
else:
    so_chan = 0
    so_le = 0
    a = n
    while n > 0:
        chu_so = n % 10
        if chu_so % 2 == 0:
            so_chan += 1
        else:
            so_le += 1
        n //= 10
    print(f"{a} có {so_chan} chữ số chẵn, {so_le} chữ số lẻ")
```

#### Bài 23
#### Nhập n
#### Cho S(k) = 1 + 2 + 3 + … + k
#### Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python
n = int(input("Nhập số: "))
if n <= 0:
    print("Vui lòng nhập số dương")
else:
    s = 0
    k = 0
    while s + (k + 1) < n:
        k += 1
        s += k
    print(f"Số k lớn nhất thỏa mãn tổng S(k) nhỏ hơn {n}: {k}")
```

#### Bài 24
#### Nhập vào A, tìm n nhỏ nhất sao cho 1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
a = int(input("Nhập số: "))
s = 1
n = 1
while s <= a:
    s += 1 / n
    n += 1
print(f"Số n bé nhất thỏa mãn tổng S(n) lớn hơn {a}: {n}")
```

#### Bài 25
#### Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
max_val = None
min_val = None

print("Nhập các số nguyên (nhập -1 để dừng):")

while True:
    # try:
        n = int(input("Nhập số: "))
        if n == -1:
            break
        if max_val is None:
            max_val = n
            min_val = n
        else:
            if n > max_val:
                max_val = n
            if n < min_val:
                min_val = n
    # except ValueError:
    #     print("Lỗi: Vui lòng chỉ nhập số nguyên!")

if max_val is not None:
    print(f"Số lớn nhất là: {max_val}")
    print(f"Số nhỏ nhất là: {min_val}")
else:
    print("Bạn chưa nhập số nào trước khi nhấn -1.")
```

#### Bài 26
#### Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
a = int(input("Nhập số nguyên dương a: "))
if a <= 0:
    print("Vui lòng nhập số nguyên dương")
else:
    if a == 1:
        print("Không có số thỏa mãn")
    else:
        m, n = 1, 1
        while True:
            s = m + n
            if s >= a:
                break
            else:
                m = n
                n = s
        print(n)
```

#### Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng). Dùng cách: for trên index; KHÔNG dùng hàm split(), len()

```python
s = input("Nhập vào một chuỗi: ")

# 1. Tìm độ dài chuỗi (thay thế cho hàm len())
length = 0
for char in s:
    length += 1

# 2. Đếm số từ bằng cách duyệt index
count = 0
for i in range(length):
    # Kiểm tra nếu ký tự hiện tại không phải khoảng trắng
    if s[i] != " ":
        # Nếu là ký tự đầu tiên HOẶC ký tự đứng trước nó là khoảng trắng thì đây là điểm bắt đầu của một từ mới
        if i == 0 or s[i-1] == " ":
            count += 1

print("Số lượng từ trong chuỗi là:", count)
```

#### Bài 28
#### Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi. Dùng cách: for trên index; KHÔNG dùng hàm split() 

```python
s = input("Nhập vào một chuỗi: ")

# Bước 1: Loại bỏ khoảng trắng thừa ở đầu chuỗi (nếu có)
s = s.strip()

# Bước 2: Duyệt qua từng vị trí (index) trong chuỗi
first_word = ""
found_space = False

for i in range(len(s)):
    # Nếu gặp khoảng trắng, dừng việc lấy ký tự
    if s[i] == " ":
        found_space = True
        break
    # Nếu chưa gặp khoảng trắng, cộng dồn ký tự vào biến first_word
    first_word += s[i]

print("Từ đầu tiên là:", first_word)
```

```python
a = input("Nhập vào chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa: ")
if a.isalpha():
    for i in a:
        if i.isupper():
            char = i.lower()
        else:
            char = i.upper()
        print(char, end = " ")
else:
    print("Chỉ được nhập chữ cái thường hoặc in hoa")
```

#### Bài 31
#### Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in thường, bao nhiêu ký tự số

```python
a = input("Nhập vao 1 chuỗi: ")
ky_tu_in_hoa = 0
ky_tu_in_thuong = 0
ky_tu_so = 0
for i in a:
    if i.isupper():
        ky_tu_in_hoa += 1
    elif i.islower():
        ky_tu_in_thuong += 1
    elif i.isnumeric():
        ky_tu_so += 1
print(f"Chuỗi có {ky_tu_in_hoa} ký tự in hoa, {ky_tu_in_thuong} ký tự in thường, {ky_tu_so} ký tự số")
```
