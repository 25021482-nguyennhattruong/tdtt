#### Bài 16
#### Nhập vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma ̣trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

```python
def print_matrix():
    try:
        # Bước 1: Nhập kích thước ma trận
        n = int(input("Nhập số hàng n: "))
        m = int(input("Nhập số cột m: "))

        # Bước 2: Nhập mảng hai chiều
        matrix = []
        print(f"Nhập các phần tử của ma trận {n}x{m}:")
        for i in range(n):
            # Nhập một dòng, tách các số bằng khoảng trắng và chuyển sang kiểu int
            row = list(map(int, input(f"Nhập dòng {i+1}: ").split()))
            
            # Kiểm tra nếu người dùng nhập thiếu hoặc thừa cột
            if len(row) != m:
                print(f"Cảnh báo: Bạn phải nhập đúng {m} phần tử. Vui lòng nhập lại dòng này.")
                # (Có thể bổ sung logic nhập lại ở đây nếu cần)
            
            matrix.append(row)

        # Bước 3: In ma trận với định dạng căn phải 4 vị trí
        print("\nMa trận vừa nhập là:")
        for row in matrix:
            for element in row:
                # Sử dụng f-string với định dạng :4d để dành 4 khoảng trống, căn phải
                print(f"{element:4d}", end="")
            print() # Xuống dòng sau khi in hết một hàng

    except ValueError:
        print("Lỗi: Vui lòng chỉ nhập số nguyên.")

# Chạy chương trình
print_matrix()
```

#### Bài 17
#### Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và đường chéo phụ của mảng

```python
# Nhập cấp của ma trận
n = int(input("Nhập cấp của ma trận vuông n: "))

# Nhập mảng 2 chiều
matrix = []
print(f"Nhập các phần tử cho ma trận {n}x{n}:")
for i in range(n):
    row = list(map(int, input(f"Nhập hàng {i+1} (cách nhau bởi dấu cách): ").split()))
    matrix.append(row)

# Lấy đường chéo chính
# Sử dụng List Comprehension để code ngắn gọn hơn
duong_cheo_chinh = [matrix[i][i] for i in range(n)]

# Lấy đường chéo phụ
duong_cheo_phu = [matrix[i][n - 1 - i] for i in range(n)]

# In kết quả
print("\nMa trận vừa nhập:")
for row in matrix:
    print(row)

print(f"\nĐường chéo chính: {duong_cheo_chinh}")
print(f"Đường chéo phụ: {duong_cheo_phu}")
```

#### Bài 18
#### Nhập vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên ̣ cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột k (hàng k)

```python
# Nhập n, m, k trên cùng một dòng
n, m, k = map(int, input("Nhập n, m, k: ").split())

# Nhập mảng 2 chiều
matrix = []
print(f"Nhập ma trận {n}x{m}:")
for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

# 1. Tính tổng các phần tử trên hàng k
if k < n:
    tong_hang_k = sum(matrix[k])
    print(f"\nTổng các phần tử trên hàng {k}: {tong_hang_k}")
else:
    print(f"\nChỉ số hàng {k} nằm ngoài phạm vi mảng.")

# 2. Tính tổng các phần tử trên cột k
if k < m:
    tong_cot_k = sum(matrix[i][k] for i in range(n))
    print(f"Tổng các phần tử trên cột {k}: {tong_cot_k}")
else:
    print(f"Chỉ số cột {k} nằm ngoài phạm vi mảng.")
```

#### Bài 20
#### Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về -1

```python
def vi_tri(a, k):
    for i in a:
        if i == k:
            return a.index(i)
    return -1

a = list(map(int, input().split()))
k = int(input())
print(vi_tri(a, k))
```

#### Bài 21
#### Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

```python
# Nhập dòng các chuỗi và tách thành danh sách
input_str = input("Nhập chuỗi: ")
words = input_str.split()

# Tạo dictionary để lưu vị trí
word_map = {}

for index, word in enumerate(words):
    if word not in word_map:
        # Nếu từ chưa có trong dict, khởi tạo bằng một list chứa vị trí hiện tại
        word_map[word] = [index]
    else:
        # Nếu từ đã có, thêm vị trí mới vào list
        word_map[word].append(index)

# Chuyển đổi list các vị trí thành tuple theo yêu cầu
result = {word: tuple(indices) for word, indices in word_map.items()}

# In kết quả
print(result)
```

#### Bài 23

```python
def xoa_nho_hon_x(arr, x):
    # Chỉ giữ lại các phần tử e nếu e >= x
    return [e for e in arr if e >= x]

nums = list(map(float, input().split()))
x = float(input("Nhập số x: "))
print(xoa_nho_hon_x(nums, x))
```

#### Bài 24
#### Viết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9

```python
def day_so_zero(arr, vi_tri):
    # Tạo danh sách chỉ chứa các số 0
    zeros = [x for x in arr if x == 0]
    # Tạo danh sách các số khác 0 (giữ nguyên thứ tự)
    others = [x for x in arr if x != 0]
    
    if vi_tri == "dau":
        return zeros + others
    else:
        return others + zeros

nums = list(map(int, input().split()))
print("Mảng gốc:  ", nums)
print("Về đầu:    ", day_so_zero(nums, "dau"))
print("Về cuối:   ", day_so_zero(nums, "cuoi"))
```

#### Bài 26
#### Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2 phần tử đầu và cuối

```python
def tong_cuc_dai_dia_phuong(arr):
    n = len(arr)
    tong = 0
    danh_sach_cuc_dai = [] # Để theo dõi các số tìm được

    # Duyệt từ phần tử thứ 2 (index 1) đến kế cuối (index n-2)
    for i in range(1, n - 1):
        if arr[i] > arr[i-1] and arr[i] > arr[i+1]:
            tong += arr[i]
            danh_sach_cuc_dai.append(arr[i])
            
    return tong, danh_sach_cuc_dai

# Nhập dữ liệu
try:
    nums = list(map(int, input("Nhập mảng các số nguyên (cách nhau bởi dấu cách): ").split()))

    if len(nums) < 3:
        print("Mảng quá ngắn để có cực đại địa phương (cần ít nhất 3 phần tử).")
    else:
        ket_qua, chi_tiet = tong_cuc_dai_dia_phuong(nums)
        print(f"Các phần tử cực đại tìm được: {chi_tiet}")
        print(f"Tổng của chúng là: {ket_qua}")
except ValueError:
    print("Vui lòng chỉ nhập số nguyên.")
```

#### Bài 27
#### Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước). nếu không có in ra -1

```python
def vi_tri(nums, x):
    list = []
    for i in nums:
        if i < x:
            list.append(i)
    if len(list) != 0:
        return list
    return -1

nums = list(map(int, input().split()))
x = int(input())
print(vi_tri(nums, x))
```

#### Bài 28
#### Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm thực sự, hay vô danh (không tăng, không giảm)

```python
def kiem_tra_day_so(arr):
    n = len(arr)
    if n < 2:
        return "Dãy quá ngắn để kết luận (cần ít nhất 2 phần tử)"

    # Kiểm tra tăng thực sự: a[i] < a[i+1] với mọi i
    la_tang = all(arr[i] < arr[i+1] for i in range(n - 1))
    
    # Kiểm tra giảm thực sự: a[i] > a[i+1] với mọi i
    la_giam = all(arr[i] > arr[i+1] for i in range(n - 1))

    if la_tang:
        return "Dãy tăng thực sự"
    elif la_giam:
        return "Dãy giảm thực sự"
    else:
        return "Dãy vô danh"

# Nhập dữ liệu và kiểm tra
try:
    nums = list(map(int, input("Nhập dãy số (cách nhau bởi dấu cách): ").split()))
    print(f"Kết quả: {kiem_tra_day_so(nums)}")
except ValueError:
    print("Vui lòng chỉ nhập số nguyên.")
```
