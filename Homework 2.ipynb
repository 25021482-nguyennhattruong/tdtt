#### Bài 1
#### Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diện tích và chu vi của nó. 

```python
a, b = map(float, input("Nhập vào 2 cạnh của 1 tam giác (mỗi số cách nhau 1 dấu cách): ").split())
dien_tich = a * b
chu_vi = (a + b) * 2
print(f"Diện tích: {dien_tich}, Chu vi: {chu_vi}")
```

#### Bài 3
#### Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Không là ba cạnh của một tam giác'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn hình.

```python
a, b, c = map(int, input("Nhập vào 3 số nguyên dương (mỗi số cách nhau 1 dấu cách): ").split())
if a > 0 and b > 0 and c > 0:
    if a + b > c and b + c > a and a + c > b:
        chu_vi = (a + b + c) / 2
        dien_tich = (chu_vi * (chu_vi - a) * (chu_vi - b) * (chu_vi - c)) ** 0.5
        print(f"Diện tích tam giác: {dien_tich:.2f}, chu vi tam giác: {chu_vi:.2f}")
        if a == b == c:
            print("Tam giác đều")
        elif a == b or b == c or c == a:
            if (a ** 2) + (b ** 2) == (c ** 2) or (b ** 2) + (c ** 2) == (a ** 2) or (a ** 2) + (c ** 2) == (b ** 2):
                print("Tam giác vuông cân")
            else:
                print("Tam giác cân")
        elif (a ** 2) + (b ** 2) == (c ** 2) or (b ** 2) + (c ** 2) == (a ** 2) or (a ** 2) + (c ** 2) == (b ** 2):
            print("Tam giác vuông")
        else:
            print("Tam giác thường")
    else:
        print("Không là ba cạnh của một tam giác")
else:
    print("Chỉ nhập số nguyên dương")
```

#### Bài 5
#### Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a, b, c = map(float, input("Nhập hệ số a, b, c (mỗi số cách nhau 1 dấu cách): ").split())
if a != 0:
    deta = (b ** 2) - (4 * a * c)
    if deta > 0:
        x_1 = (-b + (deta ** 0.5)) / (2 * a)
        x_2 = (-b - (deta ** 0.5)) / (2 * a)
        print(f"Phương trình có 2 nghiệm là {x_1:.2f} và {x_2:.2f}")
    elif deta == 0:
        print(f"Phương trình có nghiệm kép là {(-b / (2 * a)):.2f}")
    else:
        print("Phương trình vô nghiệm thực")
else:
    if b == 0:
        if c == 0:
            print("Phương trình có vô số nghiệm")
        else:
            print("Phương trình vô nghiệm")
    else:
        print(f"Phương trình có nghiệm duy nhất là {(-c / b):.2f} ")
```

#### Bài 9
#### Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao nhiêu giây. 

```python
x = int(input("Thời gian hoàn thành 1 công việc (giây): "))
if x >= 60:
    phut = x // 60
    gio = phut // 3600
    print(f"Thời gian hoàn thành 1 công việc: {gio} giờ {phut} phut {(phut * 60 - x)} giây")
elif x <= 0:
    print("Thời gian hoàn thành 1 công việc phải là số nguyên dương")
else:
    print(f"Thời gian hoàn thành 1 công việc: 0 giờ 0 phut {x} giây")
```

#### Bài 10
#### Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
a, b = map(float, input("Nhập tọa độ tâm đương tròn (mỗi số cách nhau 1 dấu cách): ").split())
r = float(input("Nhập bán kính đường tròn:"))
x, y = map(float, input("Nhập tọa độ điểm A (mỗi số cách nhau 1 dấu cách): ").split())
khoang_cach = (((a + x) ** 2) + ((b + y) ** 2))
if r == khoang_cach:
    print("Điểm A thuộc đường tròn")
else:
    print("Điểm A không thuộc đường tròn")
```
