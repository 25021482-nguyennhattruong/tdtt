#### Bài 4
#### Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím 

```python
h = int(input("Nhập vào độ cao h: "))

for i in range(1, h + 1):
    for j in range(1, 2 * h):
        # Điều kiện để in dấu sao:
        # 1. Dòng cuối cùng (i == h)
        # 2. Cạnh trái (j == h - i + 1)
        # 3. Cạnh phải (j == h + i - 1)
        if i == h or j == h - i + 1 or j == h + i - 1:
            print("*", end="")
        else:
            print(" ", end="")
    # Nhảy xuống dòng tiếp theo
    print()
```

```python
while True:
    try:
        a = int(input("Nhập vào số nguyên a: "))
        
        if a > 0:
            print("Bạn nhập đúng quy tắc")
            break
        else:
            print("Số bạn nhập không phải số dương. Vui lòng nhập lại!")
            
    except ValueError:
        print("Lỗi! Vui lòng nhập một số nguyên hợp lệ.")
```

#### Bài 12
#### Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n 

```python
n = int(input("Nhập số nguyên dương n: "))

s = 0
k = 0

while s + (k + 1) < n:
    k += 1
    s += k

print(f"Giá trị k lớn nhất là: {k}")
print(f"Tổng S({k}) = {s} < {n}")
```

#### Bài 13
#### Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho 1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 

```python
a = int(input("Nhập số nguyên dương A: "))
s = 0
n = 0

while s <= a:
    n += 1
    s += 1 / n

print(f"Giá trị n nhỏ nhất là: {n}")
print(f"Tổng S({n}) ≈ {s:.2f} > {a}")
```

#### Bài 14
#### Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng 

```python
# Khởi tạo hai biến để giữ giá trị cực đại và cực tiểu
max_val = None
min_val = None

print("Nhập các số nguyên (nhập -1 để dừng):")

while True:
    try:
        n = int(input("Nhập số: "))
        
        # Điều kiện dừng: Nếu nhập -1 thì thoát vòng lặp ngay
        if n == -1:
            break
        
        # Nếu là số đầu tiên được nhập (max_val đang là None)
        if max_val is None:
            max_val = n
            min_val = n
        else:
            # Cập nhật max nếu số mới lớn hơn số cũ
            if n > max_val:
                max_val = n
            # Cập nhật min nếu số mới nhỏ hơn số cũ
            if n < min_val:
                min_val = n
                
    except ValueError:
        print("Lỗi: Vui lòng chỉ nhập số nguyên!")

# Sau khi thoát vòng lặp, kiểm tra xem người dùng có nhập số nào hợp lệ không
if max_val is not None:
    print(f"Số lớn nhất là: {max_val}")
    print(f"Số nhỏ nhất là: {min_val}")
else:
    print("Bạn chưa nhập số nào trước khi nhấn -1.")
```

#### Bài 15
#### Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số. Làm bằng nhiều cách nhất có thể, ko dùng mảng. 

```python
n = int(input("Nhập số nguyên dương n: "))
if n < 0:
    print("Vui lòng chỉ nhập số nguyên dương")
else:
    a = n
    so = 0
    while a > 0:
        a //= 10
        so += 1
    print(f"{n} có {so} chữ số")
```

#### Bài 16
#### Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input("Nhập vào số nguyên dương n: "))
if n < 0:
    print("Vui lòng chỉ nhập số nguyên dương")
else:
    so_chan = 0
    so_le = 0
    a = n
    if n == 0:
        so_chan = 1
    else:
        while a > 0:
            chu_so = a % 10
            if chu_so % 2 == 0:
                so_chan += 1
            else:
                so_le += 1
            a //= 10
        print(f"{n} có {so_chan} số chẵn, {so_le} số lẻ")
```

#### Bài 19
#### Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
a = int(input("Nhập số nguyên dương a: "))
if a <= 0:
    print("Vui lòng nhập số nguyên dương")
else:
    if a == 1:
        print("Không có số thỏa mãn")
    else:
        m, n = 1, 1
        while True:
            s = m + n
            if s >= a:
                break
            else:
                m = n
                n = s
        print(n)
```

#### Bài 20
#### Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố

```python
def count_prime_digits(n):
    # Tập hợp các chữ số nguyên tố để kiểm tra nhanh
    prime_digits = {'2', '3', '5', '7'}
    count = 0
    
    # Chuyển n thành chuỗi để duyệt từng ký tự (chữ số)
    s = str(n) 
    
    for digit in s:
        if digit in prime_digits:
            count += 1
    return count

try:
    n = int(input("Nhập số nguyên n: "))
    if n <= 0:
        print("Vui lòng nhập số nguyên dương")
    else:
        result = count_prime_digits(n)
        print(f"Số {n} có {result} chữ số là số nguyên tố.") 
except ValueError:
    print("Chỉ được nhập các ký tự số")
```
